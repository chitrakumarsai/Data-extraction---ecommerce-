```python
def api_script():
    response = requests.get(api_url_base, headers=headers)
    shipment = response.json()
    result = pd.DataFrame()
    data_f = shipments(shipment)
    result = result.append(data_f,ignore_index=True)
    while shipment['data']['next'] is not None:
        print("Next page found, downloading", shipment['data']['next'])
        response = requests.get(shipment['data']['next'],headers=headers)
        shipment = response.json()
        data_f = shipments(shipment)
        result = result.append(data_f,ignore_index=True)
    result.to_excel('FinalOutput_2.xlsx', index=False)
    today = date.today()
    print("Today's date:", today)
```

```python
schedule.every().day.at("08:30").do(api_script)
```
